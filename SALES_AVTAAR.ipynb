```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sales_full_data.csv")
```

```python
df
```

```python
df.isnull()
```

```python
df = df.dropna()
df
```

```python
df.isnull()
```

```python
df.dtypes
```

```python
list(df.columns)
```

```python
df = df[df['Quantity Ordered'] != 'Quantity Ordered']
```

```python
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])
```

```python
df['Price Each'] = pd.to_numeric(df['Price Each'])
```

```python
df['Sales'] = df['Quantity Ordered'] * df['Price Each']
df
```

```python
df['Month'] = df['Order Date'].str[0:2]
df['Month'] = df['Month'].astype('int32')
df.head()
```

```python
mon = df.groupby('Month').sum()
mon
```

```python
months = range(1,13)
plt.bar(months,mon['Sales'])
plt.xticks(months)
plt.ylabel('Sale in USD $')
plt.xlabel('Months')
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df["Hour"] = df['Order Date'].dt.hour
df["Minute"] = df['Order Date'].dt.minute
df.head()
```

```python
hours = df.groupby('Hour').sum()
hours
```

```python
hours = [hour for hour, df in df.groupby('Hour')]
plt.plot(hours,df.groupby(['Hour']).count())
plt.xticks(hours)
plt.xlabel('Hours')
plt.ylabel('Number of orders')
```

```python
data = []

for Product in df['Product'].unique():
    data.append([Product , len(df[df['Product']  == Product])])

pd.DataFrame(data,columns = ['Product','No. of Product']).sort_values(by='No. of Product', ascending=False).plot(x = 'Product', y = 'No. of Product', kind = 'bar', figsize = (10,5))
```
